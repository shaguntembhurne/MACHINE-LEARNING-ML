```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Set up the environment
%matplotlib inline
%reload_ext autoreload
%autoreload

# Load and view dataset
sns.get_dataset_names()
crash_df = sns.load_dataset("car_crashes")
print(crash_df)

```

```python
# Distribution Plot
sns.displot(crash_df['no_previous'], kde=True, bins=25)
```

```python
# Joint Plot
sns.jointplot(x="alcohol", y="speeding", data=crash_df, kind="reg")
```

```python
# KDE Plot
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.kdeplot(x=crash_df['alcohol'], y=crash_df['speeding'])
```

```python
# Pair Plot
sns.pairplot(crash_df)
```

```python
# Load another dataset
tips_df = sns.load_dataset('tips')
print(tips_df)

# Pair Plot with hue
sns.pairplot(tips_df, hue="sex", palette="Blues")
```

```python
# Rug Plot
sns.rugplot(tips_df["tip"])

# Styling
sns.set_style('darkgrid')
plt.figure(figsize=(8,4))
sns.jointplot(x="alcohol", y="speeding", data=crash_df, kind="reg")
sns.set_context("paper")
```

```python

# Categorical Data Plots
sns.barplot(x="sex", y='total_bill', data=tips_df, estimator=np.std)
```

```python

# Count Plot
sns.countplot(x="sex", data=tips_df)
```

```python

# Box Plot
sns.boxplot(x="day", y='total_bill', data=tips_df, hue="sex")
```

```python

# Violin Plot
sns.violinplot(x="day", y='total_bill', data=tips_df, hue="sex")
```

```python

# Strip Plot
plt.figure(figsize=(8,5))
sns.stripplot(x="day", y='total_bill', data=tips_df, dodge=True, hue="sex")
```

```python

# Swarm Plot
sns.violinplot(x="day", y='total_bill', data=tips_df)
sns.swarmplot(x="day", y='total_bill', data=tips_df)

# Palette Styling
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x="day", y='total_bill', data=tips_df, hue="sex", palette="afmhot")
plt.legend(loc=0)

# Matrix Plot
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index="month", columns="year", values="passengers")
sns.heatmap(flights, cmap="Blues", linecolor="white", linewidths=1)

# PairGrid on Iris Dataset
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue="species")
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)

# FacetGrid
g = sns.FacetGrid(crash_df, col="speeding", col_wrap=3, height=4)
g.map(sns.histplot, "alcohol", kde=True)
```
