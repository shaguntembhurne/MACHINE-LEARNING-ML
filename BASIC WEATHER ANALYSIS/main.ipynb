```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Creating a dummy weather data
data = {
    "date": pd.date_range(start="2024-11-01", periods=10),
    "Temperature (C)": np.random.randint(15, 30, 10),
    "Humidity (%)": np.random.randint(30, 90, 10)
}

# Creating the dataframe
df = pd.DataFrame(data)

# Calculate mean temperature and mean humidity
mean_temp = np.mean(df["Temperature (C)"])
mean_humidity = np.mean(df["Humidity (%)"])

# Plotting the Temperature trend
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df["Temperature (C)"], label="Temperature trend", color="red", marker="o")
plt.title('Temperature Trend')
plt.grid(True)
plt.xlabel("Date")
plt.ylabel('Temperature (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()

# Bar graph for humidity analysis
plt.figure(figsize=(10, 6))
plt.bar(df['date'], df['Humidity (%)'])
plt.xlabel("Date")
plt.ylabel('Humidity (%)')
plt.title('Humidity Analysis')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()
```
